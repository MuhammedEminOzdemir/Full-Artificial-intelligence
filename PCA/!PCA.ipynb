## PCA

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris = load_iris()
data = iris.data
feature_names = iris.feature_names
y = iris.target
```

```python
len(feature_names)
```

```python
y
```

```python
df = pd.DataFrame(data, columns=feature_names)
df["sinif"] = y
x = data
df.head(20)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=4, whiten=True) 
```

```python
pca.fit(x)
```

```python
x_pca = pca.transform(x)
```

```python
print(x.shape)
print(x_pca.shape)
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
df_pca_c = pd.DataFrame({'variance' : pca.explained_variance_ratio_,
                        'PC' : ['PC_'+ str(i) for i in range(1,x_pca.shape[1]+1)]})
df_pca_c
```

```python
import seaborn as sns
sns.barplot(x='PC',y='variance',data=df_pca_c,color='c')
plt.ylabel('Variance Explained')
plt.xlabel('Principle Components')
plt.show()
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
r = [i for i in range(1, 5)]
plt.plot(r, np.cumsum(pca.explained_variance_ratio_))
```

```python
# 2D
df["p1"] = x_pca[:,0]
df["p2"] = x_pca[:,1]

color = ["red","green","blue"]

import matplotlib.pyplot as plt
for each in range(3):
    plt.scatter(df.p1[df.sinif == each],df.p2[df.sinif == each],color = color[each],label = iris.target_names[each])
    
plt.legend()
plt.xlabel("p1")
plt.ylabel("p2")
plt.show()
```

## dimension reduction problem

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

```python
pca = PCA(n_components = 4)  # project from 64 to 2 dimensions
projected = pca.fit_transform(digits.data)
print(digits.data.shape)
print(projected.shape)
```

```python
len(pca.components_)
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
pca.n_components
```

```python
pca.singular_values_
```

```python
plt.figure(figsize=(12,8))
plt.scatter(projected[:, 0], projected[:, 1],
            c=digits.target, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('Spectral', 10))
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

```python
pca = PCA().fit(digits.data)
plt.figure(figsize=(18,6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.axhline(0.90, c="r")
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```
